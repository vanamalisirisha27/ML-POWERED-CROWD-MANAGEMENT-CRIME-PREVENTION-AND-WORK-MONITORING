```python
import os
from retinaface import RetinaFace
import numpy as np
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt

# Load pre-trained RetinaFace model
retinaface = RetinaFace.build_model()

# Path to the preprocessed LFW dataset
dataset_dir = r'C:/crowdcrime/yolov8-streamlit-detection-tracking/face/preprocessed_custom_dataset'
subsets = ['train', 'test', 'val']

# Output directory for aligned faces
output_dir = r'C:/crowdcrime/yolov8-streamlit-detection-tracking/face/aligned_faces'
os.makedirs(output_dir, exist_ok=True)

# Iterate through each subset
for subset in subsets:
    subset_dir = os.path.normpath(os.path.join(dataset_dir, subset))
    output_subset_dir = os.path.normpath(os.path.join(output_dir, subset))
    os.makedirs(output_subset_dir, exist_ok=True)
    
    for person_name in os.listdir(subset_dir):
        person_dir = os.path.normpath(os.path.join(subset_dir, person_name))
        output_person_dir = os.path.normpath(os.path.join(output_subset_dir, person_name))
        os.makedirs(output_person_dir, exist_ok=True)
        
        if os.path.isdir(person_dir):
            for img_name in os.listdir(person_dir):
                img_path = os.path.normpath(os.path.join(person_dir, img_name))

                # Load input image
                image = cv2.imread(img_path)

                # Perform face detection with RetinaFace
                faces = RetinaFace.detect_faces(img_path=img_path)

                # Iterate over detected faces
                for idx, face in enumerate(faces.values()):
                    # Extract bounding box coordinates
                    bbox = face["facial_area"]
                    x1, y1, x2, y2 = bbox

                    # Crop the face region from the original image
                    face_region = image[y1:y2, x1:x2]

                    # Save aligned face to the output directory
                    output_path = os.path.join(output_person_dir, f'{person_name}{img_name}{idx}.jpg')
                    cv2.imwrite(output_path, face_region)

                    # Print detected and aligned faces
                    print(f"Detected and aligned face in {img_path}")

                    # Display image using Matplotlib
                    #plt.imshow(cv2.cvtColor(face_region, cv2.COLOR_BGR2RGB))
                    #plt.axis('off')
                    #plt.show()

print("Face alignment completed.")
```
